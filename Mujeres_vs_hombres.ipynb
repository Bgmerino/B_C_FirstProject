```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('./Gender_Stats_csv/Gender_StatsData.csv')
df.head()
```

```python
#df.index=df['Country Code']
#dfClean = df.drop(['Country Code'], axis=1)
dfClean = df.drop(['Country Name'],axis=1)
dfCleanv2 = dfClean.drop(['Indicator Code'], axis=1)
dfCleanv2.head()
```

```python
print(dfCleanv2['Country Code'].unique(),"\n",dfCleanv2['Indicator Name'].unique())
Pais='ARB'
Ind_name= 'Access to anti-retroviral drugs'
```

```python
print(Ind_name)
dfSelect=dfCleanv2[dfCleanv2['Indicator Name'].str.contains(Ind_name)]
dfSelect.head()
```

```python
dfSelect_male=dfSelect[dfSelect['Indicator Name'].str.contains(' male')]
dfSelect_maleC=dfSelect_male.drop('Indicator Name',axis=1)
dfSelect_maleC = dfSelect_maleC.set_index('Country Code')

dfSelect_female=dfSelect[dfSelect['Indicator Name'].str.contains('female')]
dfSelect_femaleC=dfSelect_female.drop('Indicator Name',axis=1)        
dfSelect_femaleC = dfSelect_femaleC.set_index('Country Code')

dfSelect_femaleC.head()
```

```python
tt=dfSelect_femaleC.T
tt.index
```

```python
new_index=[]
for indice in tt.index:
    try:
        new_index.append(int(indice))
    except:
        new_index.append(2017)
    
tt.index=new_index
print(tt.index)
                     
```

```python
ax=tt.plot(figsize=(15,15))
ax.set_xlim(2000,2016)
plt.show()
```

```python
tt.head()
```

```python
plt.figure(figsize=(8,5), dpi=100)
paises=[]
for keys in tt:
    if np.any(tt[keys].values > 70):
        
        plt.plot(tt[keys].index,tt[keys].values,label=keys)
        paises.append(keys)
        
plt.legend(loc='upper left', frameon=False)
plt.show()
```

```python
mm=dfSelect_maleC.T
mm.index
new_index=[]
for indice in mm.index:
    try:
        new_index.append(int(indice))
    except:
        new_index.append(2017)
    
mm.index=new_index

plt.figure(figsize=(8,5), dpi=100)
paises=[]
for keys in mm:        
      if np.any(mm[keys].values > 70):
        plt.plot(mm[keys].index,mm[keys].values,label=keys)
        paises.append(keys)
        
plt.legend(loc='upper left', frameon=False)
plt.show()
                     
```
