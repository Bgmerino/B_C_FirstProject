```python
# dfClean.iloc[3:5][1:2]
dfCleanTrans=dfClean.T
dfCleanTrans.head()
```

```python
dfCleanTrans['Albania']
```

```python
plt.figure(figsize=(8,5), dpi=100)
paises=[]
for keys in dfCleanTrans:
     if np.any(dfCleanTrans[keys].values > 40):
            #dfCleanTrans[keys].plot(figsize=(15,15))
            plt.plot(dfCleanTrans[keys].index,dfCleanTrans[keys].values,label=keys,alpha=0.5)
            size=[abs(number) for number in dfCleanTrans[keys].values]
            plt.scatter(dfCleanTrans[keys].index,dfCleanTrans[keys].values,s=size,alpha=0.5)
            paises.append(keys)
            
plt.legend(loc='upper left', frameon=False)
plt.show()
```
